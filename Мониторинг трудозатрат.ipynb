```python
def isNaN(num):
    return num != num
```

```python
def add_time(time1, time2):
    if (type(time1) != int):
        secs1 = time1.hour * 3600 + time1.minute * 60 + time1.second
        secs2 = time2.hour * 3600 + time2.minute * 60 + time2.second
        secs = secs1 + secs2
    else:
        secs2 = time2.hour * 3600 + time2.minute * 60 + time2.second
        secs = time1 + secs2
    return secs
```

```python
def get_time(secs, total_time = False):
    
    hours = secs // 3600
    minutes = (secs % 3600) // 60
    seconds = secs % 60
            
    if (total_time == False):
        if seconds >= 30:
            minutes += 1
        if minutes >= 30:
            hours += 1
        return hours
    else:
        return dt.time(hour = hours, minute = minutes, second = seconds)
```

```python
task51 = [[]]
task53 = [[]]

str51 = '5.1'
str53 = '5.3'
```

```python
errors51 = pd.DataFrame(columns=['ID', 'Название', 'Исполнитель', 'Подразделение', 'Дата создания', 'Категория специалиста', 'Затраченное время детально', 'Комментарий', 'Ошибка время', 'Ошибка комментарий'])
errors53 = pd.DataFrame(columns=['ID', 'Название', 'Исполнитель', 'Подразделение', 'Дата создания', 'Категория специалиста', 'Затраченное время детально', 'Комментарий', 'Ошибка время', 'Ошибка комментарий'])
```

```python
errors_total_time = pd.DataFrame(columns=['ID', 'Название', 'Исполнитель', 'Подразделение', 'Дата создания', 'Категория специалиста', 'Суточное время'])
```

```python
for i in range(len(df)):
    # Checking the date of the submission
    date = dt.datetime.strptime(df['Дата создания'][i], '%Y-%m-%d')
    if ((date <= time_range[1]) and (date >= time_range[0])):
        # Filling the array for task 5.1
        if (df['Название'][i].find(str51) != (-1)):
            
            if (df['Исполнитель'][i] in task51[0]) == False:
                task51[0].append(df['Исполнитель'][i])
                dic = {'worker': df['Исполнитель'][i], 'department': df['Подразделение'][i], 'total_time': 0, 'total_correct_time': 0, 'total_correct_comment_time': 0, 'mistake_time': 0, 'mistake_comment': 0, 'mistake_total_time' : 0}
                task51.append(dic)
            if (isNaN(df['Затраченное время детально'][i]) == False):
                
                # Adding to total time
                task51[task51[0].index(df['Исполнитель'][i]) + 1]['total_time'] = add_time(task51[task51[0].index(df['Исполнитель'][i]) + 1]['total_time'], df['Затраченное время детально'][i]) 

                # Adding to correct total time 
                if (type(df['Затраченное время детально'][i]) == dt.datetime):
                    df.loc[i, ['Затраченное время детально']] = (df['Затраченное время детально'][i]).time()
                    
                if (df['Затраченное время детально'][i] >= dt.time(0,5,0)) & (df['Затраченное время детально'][i] <= dt.time(4,0,0)):  
                    task51[task51[0].index(df['Исполнитель'][i]) + 1]['total_correct_time'] = add_time(task51[task51[0].index(df['Исполнитель'][i]) + 1]['total_correct_time'], df['Затраченное время детально'][i])

                    # Adding to correct total time with valid comments
                    if (isNaN(df['Комментарий'][i]) == False):  
                        task51[task51[0].index(df['Исполнитель'][i]) + 1]['total_correct_comment_time'] = add_time(task51[task51[0].index(df['Исполнитель'][i]) + 1]['total_correct_comment_time'], df['Затраченное время детально'][i]) 
                    else: 
                        task51[task51[0].index(df['Исполнитель'][i]) + 1]['mistake_comment'] += 1

                elif (isNaN(df['Комментарий'][i]) == False):
                    task51[task51[0].index(df['Исполнитель'][i]) + 1]['mistake_time'] += 1

                    # Addind the row to the errors51 dataframe
                    errors51.loc[len(errors51.index)] = [df['ID'][i], df['Название'][i], df['Исполнитель'][i], df['Подразделение'][i], df['Дата создания'][i], df['Категория специалиста'][i], df['Затраченное время детально'][i], df['Комментарий'][i], 'Ошибка', 'Верно']
                    
                else:
                    task51[task51[0].index(df['Исполнитель'][i]) + 1]['mistake_comment'] += 1
                    task51[task51[0].index(df['Исполнитель'][i]) + 1]['mistake_time'] += 1
                    

        # Filling the array for task 5.3
        elif (df['Название'][i].find(str53) != (-1)):
            if (df['Исполнитель'][i] in task53[0]) == False:
                task53[0].append(df['Исполнитель'][i])
                dic = {'worker': df['Исполнитель'][i], 'department': df['Подразделение'][i], 'total_time': 0, 'total_correct_time': 0, 'total_correct_comment_time': 0, 'mistake_time': 0, 'mistake_comment': 0, 'mistake_total_time' : 0}
                task53.append(dic)
            if (isNaN(df['Затраченное время детально'][i]) == False):
                
                # Adding to total time
                task53[task53[0].index(df['Исполнитель'][i]) + 1]['total_time'] = add_time(task53[task53[0].index(df['Исполнитель'][i]) + 1]['total_time'], df['Затраченное время детально'][i]) 
                
                # Adding to correct total time 
                if (type(df['Затраченное время детально'][i]) == dt.datetime):
                    df.loc[i, ['Затраченное время детально']] = (df['Затраченное время детально'][i]).time()

                if (df['Затраченное время детально'][i] >= dt.time(0,5,0)) & (df['Затраченное время детально'][i] <= dt.time(4,0,0)):  
                    task53[task53[0].index(df['Исполнитель'][i]) + 1]['total_correct_time'] = add_time(task53[task53[0].index(df['Исполнитель'][i]) + 1]['total_correct_time'], df['Затраченное время детально'][i])

                    # Adding to correct total time with valid comments
                    if (isNaN(df['Комментарий'][i]) == False):  
                        task53[task53[0].index(df['Исполнитель'][i]) + 1]['total_correct_comment_time'] = add_time(task53[task53[0].index(df['Исполнитель'][i]) + 1]['total_correct_comment_time'], df['Затраченное время детально'][i]) 
                    else: 
                        task53[task53[0].index(df['Исполнитель'][i]) + 1]['mistake_comment'] += 1

                elif (isNaN(df['Комментарий'][i]) == False):
                    task53[task53[0].index(df['Исполнитель'][i]) + 1]['mistake_time'] += 1
                        
                else:
                    task53[task53[0].index(df['Исполнитель'][i]) + 1]['mistake_comment'] += 1
                    task53[task53[0].index(df['Исполнитель'][i]) + 1]['mistake_time'] += 1
```
